Для видалення шуму імпульсного характеру оптимальним способом є застосування медіанного фільтра.  
При медіанній фільтрації кожен піксель зображення оброблюється послідовно. Використовується двовимірне вікно (апертура; моя програма використовує вікно у вигляді квадрату, розмір вікна задається параметрично).  
Для послідовності точок, що належать вікну знаходимо медіану (впорядковуємо точки і обираємо середню).  
Знайдена медіана і буде значенням точки зображення з координатами (i, j).

Код програми:

```python
import numpy as np
from PIL import Image


def median_filter(image_arr, window_size):
    m, n = image_arr.shape
    result_image_arr = np.zeros((m, n))
    h = (window_size - 1) // 2
    
    for i in range(m):
        for j in range(n):
            window = image_arr[max(i - h, 0):min(i + h + 1, m), max(j - h, 0):min(j + h + 1, n)]
            result_image_arr[i, j] = np.median(window)
            
    return result_image_arr.astype(np.uint8)
```

Допоміжна функція для виводу зображення:

```python
def show_image(image):
    target_width = 350
    target_height = int((target_width / image.width) * image.height)
    return image.resize((target_width, target_height))
```

Зображення Saturn2.gif (початкове):

```python
first_image = Image.open('Saturn2.gif').convert('L')
show_image(first_image)
```

Зображення Saturn2.gif (після застосування медіанної фільтрації з вікном 3 х 3):

```python
show_image(Image.fromarray(median_filter(np.array(first_image), 3)))
```

Зображення saturn3.GIF (початкове):

```python
second_image = Image.open('saturn3.GIF').convert('L')
show_image(second_image)
```

Зображення saturn3.GIF (після застосування медіанної фільтрації з вікном 5 х 5):

```python
show_image(Image.fromarray(median_filter(np.array(second_image), 5)))
```
